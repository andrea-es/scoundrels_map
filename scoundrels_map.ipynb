#### Locate Markers

Get markers pixels

```python
# Wait for map elements to load
WebDriverWait(driver, 10).until(
    EC.presence_of_all_elements_located((By.CLASS_NAME, "leaflet-marker-icon"))
)

# Find all map elements with the specific class
map_elements = driver.find_elements(By.CLASS_NAME, "leaflet-marker-icon")

# Create an empty list to store element details
data = []

# Extract details of each element and store them in the list
for index, element in enumerate(map_elements, start=1):
    element_class = element.get_attribute("class")
    
    # Skip rows containing user icons
    if "wm-user-icon" in element_class:
        continue
    
    element_location = element.location  # Element's position on the map
    
    # Extract event type from class string
    event_type = None
    for cls in element_class.split():
        if "wm-alert-icon--" in cls and cls != "wm-alert-icon":
            event_type = cls.replace("wm-alert-icon--", "")
            break  # Stop searching once the event type is found
    
    # Additional check for "cluster" or other hints in the class
    if not event_type:
        if "current-location" in element_class:
            event_type = "current-location"
        elif "is-origin" in element_class:
            event_type = "origin"
        elif "is-destination" in element_class:
            event_type = "destination"
        elif "road-closed" in element_class:
            event_type = "road-closed"
        elif "police" in element_class:
            event_type = "police"
        elif "hazard" in element_class:
            event_type = "hazard"
        elif "accident" in element_class:
            event_type = "accident"
        elif "wm-alert-cluster-icon" in element_class:
            event_type = "cluster"
    
    # Append the extracted data to the list
    data.append({
        "Event Type": event_type,
        "Class": element_class,
        "Location_px": element_location
    })

# Create a pandas DataFrame from the list
df = pd.DataFrame(data)
```

Get coordinates of each point

```python
# Known pixel coordinates for current location and destination
current_px = df.loc[df["Event Type"] == "current-location", "Location_px"].values[0]
destination_px = df.loc[df["Event Type"] == "destination", "Location_px"].values[0]

# Calculate scaling factors
lat_scale = (destination_latitude - current_latitude) / (destination_px["y"] - current_px["y"])
lon_scale = (destination_longitude - current_longitude) / (destination_px["x"] - current_px["x"])

# Iterate through the DataFrame
df["Location_coord"] = None  # Initialize the column

for index, row in df.iterrows():
    if row["Event Type"] == "current-location":
        df.at[index, "Location_coord"] = {"lat": current_latitude, "lon": current_longitude}
    elif row["Event Type"] == "destination":
        df.at[index, "Location_coord"] = {"lat": destination_latitude, "lon": destination_longitude}
    else:
        px = row["Location_px"]
        lat = current_latitude + (px["y"] - current_px["y"]) * lat_scale
        lon = current_longitude + (px["x"] - current_px["x"]) * lon_scale
        df.at[index, "Location_coord"] = {"lat": round(lat,7), "lon": round(lon,7)}
```

Create circular area to avoid traffic events

```python
from geopy.distance import geodesic

def create_circle_polygon(center, radius, num_points=10):
    """
    Create a polygon approximating a circle.
    
    Args:
        center (tuple): (latitude, longitude) of the center.
        radius (float): Radius of the circle in meters.
        num_points (int): Number of points to approximate the circle.
    
    Returns:
        list: A GeoJSON-compatible list of coordinates for the polygon.
    """
    center_lat, center_lon = center
    angle_step = 360 / num_points
    coordinates = []

    for i in range(num_points + 1):  # +1 to close the circle
        angle = i * angle_step
        # Calculate new point on the circle
        point = geodesic(kilometers=radius / 1000).destination((center_lat, center_lon), angle)
        coordinates.append([point.longitude, point.latitude])

    return [coordinates]
```

```python
# List to store areas to avoid
areas_to_avoid = []

# Radius for avoidance area (e.g., 100 meters)
avoidance_radius = 100

# Filter rows with 'road-closed' or 'police' event types
for _, row in df.iterrows():
    if row['Event Type'] in ['road-closed', 'police']:
        location_coord = row['Location_coord']
        center = (location_coord['lat'], location_coord['lon'])
        # Create a polygon for the area to avoid
        polygon = create_circle_polygon(center, avoidance_radius)
        areas_to_avoid.append(polygon)

# Output the list of areas to avoid
#areas_to_avoid
```

Get route

```python
avoid_polygons = {
    "type": "MultiPolygon",
    "coordinates": areas_to_avoid}

# Request directions avoiding the polygons
route = client.directions(
    coordinates=[start,end],
    profile="driving-car",
    format="geojson",
    options={"avoid_polygons": avoid_polygons}
)

# Print the summary of the route
summary = route['features'][0]['properties']['summary']
print(f"Total distance: {summary['distance'] / 1000:.2f} km")
print(f"Estimated time: {summary['duration'] / 60:.2f} minutes")
```

Print route

```python
import folium

# Use the coordinates directly since they are already decoded
geometry = route['features'][0]['geometry']
coordinates = geometry['coordinates']  # List of [longitude, latitude] pairs

# Create a folium map
# Center the map on the first coordinate
m = folium.Map(location=[coordinates[0][1], coordinates[0][0]], zoom_start=13)

# Add the route to the map
folium.PolyLine([(coord[1], coord[0]) for coord in coordinates], color="blue", weight=5).add_to(m)

# Save and display the map
m.save("route_map.html")
print("Map saved to 'route_map.html'. Open this file to view the route.")

```
